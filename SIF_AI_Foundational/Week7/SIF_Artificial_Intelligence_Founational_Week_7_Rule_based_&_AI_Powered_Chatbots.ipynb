#Running the Rule-based Chatbot

Rule-based chatbots operate on a set of predefined rules that dictate how they respond to user input. These rules can be as simple as a keyword matching system, where the bot looks for certain keywords in the user's input and responds accordingly, or they can be more complex, involving natural language processing and machine learning algorithms. Rule-based chatbots tend to be more deterministic than generative chatbots, as they rely on a fixed set of rules to generate responses.

#Import necessary libraries

The code imports the re and random libraries. re is used for regular expression matching and random is used to randomly select responses from the keywords dictionary.

````python
# Import necessary libraries
import re
import random

# Define a dictionary of rules and responses
rules = {
    'do you know about sustainability?': ['Sustainability is the ability to meet the needs of the present without compromising the ability of future generations to meet their own needs.'],
    'what is sustainability?': ['Sustainability is the ability to meet the needs of the present without compromising the ability of future generations to meet their own needs.'],
    'why is sustainability important?': ['Sustainability is important because it ensures that we are able to meet our needs without compromising the needs of future generations.'],
    'how can we be sustainable?': ['We can be sustainable by reducing our use of natural resources, recycling, and using renewable energy sources.'],
    'what are some examples of sustainable practices?': ['Some examples of sustainable practices include reducing water usage, using energy-efficient appliances, and using public transportation.']
}

# Define a dictionary of keywords and related responses
keywords = {
    'sustainability': ['Sustainability is the ability to meet the needs of the present without compromising the ability of future generations to meet their own needs.', 'Sustainability is important because it ensures that we are able to meet our needs without compromising the needs of future generations.', 'We can be sustainable by reducing our use of natural resources, recycling, and using renewable energy sources.'],
    'renewable energy': ['Renewable energy sources are sources of energy that can be replenished naturally, such as solar, wind, and hydro power.', 'Using renewable energy sources is an important part of being sustainable.'],
    'recycling': ['Recycling is the process of converting waste materials into new materials and objects. It is an important part of being sustainable.', 'Recycling helps to conserve natural resources and reduce the amount of waste that goes into landfills.'],
    'water usage': ['Reducing water usage is an important part of being sustainable.', 'Conserving water helps to reduce the strain on water resources and ensure that there is enough water for everyone.'],
    'public transportation': ['Using public transportation is a sustainable alternative to driving alone.', 'Taking public transportation helps to reduce traffic congestion, air pollution, and greenhouse gas emissions.']
}

# Define a function to generate a response based on a message
def generate_response(message):
    # Convert the message to lowercase
    message = message.lower()
    # Check for keywords in the message
    matched_keywords = []
    for keyword in keywords:
        if keyword in message:
            matched_keywords.append(keyword)
    # If no keywords are found, return a default response
    if not matched_keywords:
        return "I'm sorry, I don't understand. Could you please rephrase your question?"
    # Generate a response based on the matched keywords
    response = []
    for keyword in matched_keywords:
        response += random.choice(keywords[keyword])
    return ' '.join(response)

# Define a function to start the chatbot
def chat():
    print("Hi, I'm a chatbot that can help you learn about sustainability. Ask me anything!")
    while True:
        message = input("You: ")
        if message == 'bye':
            print('Chatbot: Goodbye!')
            break
        else:
            response = generate_response(message)
            print('Chatbot: ' + response)
````

````python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the pre-trained GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2', return_dict=True)

# Define a function to generate the chatbot's response
def generate_response(prompt):
  input_ids = tokenizer.encode(prompt, return_tensors='pt')
  output = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
  response = tokenizer.decode(output[0], skip_special_tokens=True)
  return response

# Start the conversation
print("Hello! I'm a chatbot. How can I help you today?")

while True:
  user_input = input("You: ")
  if user_input.lower() in ['bye', 'goodbye', 'exit', 'quit']:
    print("Chatbot: Goodbye! Have a great day.")
    break
  response = generate_response(user_input)
  print("Chatbot:", response)
````
