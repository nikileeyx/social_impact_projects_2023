##Example of Task

Loops are an essential concept in programming that allow you to repeat a block of code multiple times. In Python, there are two types of loops: for loops and while loops.

A for loop is used when you want to iterate over a sequence of values, such as a list, tuple, or string. The loop executes once for each element in the sequence, with a loop variable taking on the value of each element in turn.

A while loop is used when you want to repeat a block of code while a certain condition is true. The loop continues to execute as long as the condition is true.

Both types of loops use indentation to define the block of code that should be repeated. This block of code is often referred to as the loop body.

Here's an example of a for loop in Python:


##**Now you can try the question!**

````python
# Define a list of numbers
numbers = [1, 2, 3, 4, 5]

# Use a for loop to iterate through the list
for number in numbers:
    # Print each number
    print(number)
````

In this example, the for loop iterates through each element of the numbers list and prints it using the print() function.

Here's an example of a while loop in Python:

````python
# Define a counter variable
counter = 0

# Use a while loop to repeat a block of code while the counter is less than 5
while counter < 5:
    # Print the value of the counter
    print(counter)
    # Increment the counter by 1
    counter += 1
````

In this example, the while loop continues to execute as long as the counter variable is less than 5. The loop body prints the value of the counter using the print() function, and then increments the counter by 1 using the += operator.

#Using a "while" loop

In this example, we define a list of numbers and initialize the loop counter i to zero. 

We then use a while loop to iterate through the list while the counter is less than the length of the list. 

Inside the loop, we print the current number at index i of the list and increment the loop counter. 

The loop continues until the counter i reaches the length of the list, at which point the loop terminates.


````python
# Define a list of numbers
numbers = [1, 2, 3, 4, 5]

# Initialize the loop counter
i = 0

# Use a while loop to iterate through the list
while i < len(numbers):
    # Print the current number
    print(numbers[i])
    
    # Increment the loop counter
    i += 1
````
