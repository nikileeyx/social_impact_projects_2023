#Task 1: Understanding Dictionaries in Python

In Python, a dictionary is a collection of key-value pairs, where each key is associated with a value. Dictionaries are a built-in data type in Python and are commonly used to store and retrieve data in an efficient way.



##Here's an example of how to create a dictionary in Python:

In this example, we create a dictionary with three key-value pairs: "apple" is associated with the value 1, "banana" is associated with the value 2, and "orange" is associated with the value 3. We then print the dictionary, which outputs:


````python
# Create a dictionary
my_dict = {"apple": 1, "banana": 2, "orange": 3}

# Print the dictionary
print(my_dict)
````

##Accessing the Keys and values

To access a value in a dictionary, you can use its corresponding key. In this example, we access the value associated with the key "apple", which outputs the value of "1"



````python
# Access a value in the dictionary
print(my_dict["apple"])
````

##Updating the Keys and values

You can also add, update, or remove key-value pairs in a dictionary:

````python
# Add a new key-value pair to the dictionary
my_dict["grape"] = 4

# Update the value associated with a key
my_dict["banana"] = 5

# Remove a key-value pair from the dictionary
del my_dict["orange"]

# Print the updated dictionary
print(my_dict)
````

In this example, we add a new key-value pair "grape": 4 to the dictionary, update the value associated with the key "banana" to 5, and remove the key-value pair associated with the key "orange". The updated dictionary is then printed:



```
{'apple': 1, 'banana': 5, 'grape': 4}
```

# Task 3: Visualization of Data through Maps! üåè

In this example code, we have sample data consisting of three locations, each with a name, address, latitude, and longitude. We create a map using folium.Map(), specifying the center of the map and the zoom level.

---


We then loop through the data, extracting the latitude, longitude, name, and address information for each location. We create a popup text string for each marker, and then add a marker to the map using folium.Marker().


Finally, we display the map using m.

## Let's try with one location in the world!

We are using Central Park in New York! You can search "central park new york coordinates" on Google to find the latitude and longitude of the locations easily! üòä


"data" is a list of length 1, it contains one dictionary.

````python
# Import necessary libraries
import folium

# Sample data as a dictionary
data = [{
        "name": "Central Park",
        "address": "New York, NY",
        "latitude": 40.7829,
        "longitude": -73.9654
}]
 
# Create a map centered on a specific location
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Loop through the data and add markers to the map
for location in data:
    lat = location['latitude']
    lon = location['longitude']
    name = location['name']
    address = location['address']
    popup_text = f"<b>{name}</b><br/>{address}"
    marker = folium.Marker(location=[lat, lon], popup=popup_text)
    marker.add_to(m)

# Display the map
m
````

## Let's try with another two locations!

````python
# Import necessary libraries
import folium

# Sample data
data = [
    {
        "name": "Central Park",
        "address": "New York, NY",
        "latitude": 40.7829,
        "longitude": -73.9654
    }]

# Let's include two more locations to the dataset! We are creating a list to combine with the original list
new_locations = [{
        "name": "Golden Gate Bridge",
        "address": "San Francisco, CA",
        "latitude": 37.8199,
        "longitude": -122.4783
    },
    {
        "name": "Space Needle",
        "address": "Seattle, WA",
        "latitude": 47.6205,
        "longitude": -122.3493
    }]

#Combining the two lists of dictionaries, to form one list of dictionaries
new_data = data + new_locations

# Create a map centered on a specific location
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Loop through the data and add markers to the map
for location in new_data:
    lat = location['latitude']
    lon = location['longitude']
    name = location['name']
    address = location['address']
    popup_text = f"<b>{name}</b><br/>{address}"
    marker = folium.Marker(location=[lat, lon], popup=popup_text)
    marker.add_to(m)

# Display the map
m
````
